```
## Customer Age?
# 31-40 years=3
#21-30 years=2
#41-50 yaers=4
#51 years and above=5
#Less than 20 years=1
Age_counts=df.Age.value_counts(dropna=False)
Age_counts
plt.figure(figsize=(12,6))
plt.title("Target counts")
plt.pie(Age_counts,labels=Age_counts.index,autopct='%1.1f%%')
```
